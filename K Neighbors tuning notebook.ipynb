```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
all_games_df = pd.read_excel('Games_Database.xlsx', engine='openpyxl', header=1, nrows=256)

#all_games_df = pd.read_excel('Games_Database.xlsx', sheet_name = 'Temp DB', engine='openpyxl', header=1, nrows=86)
all_games_df.tail()
```

```python
data = all_games_df.fillna(0).drop(['Team', 'Points', 'Opponent Points', 'Team2', 'Date', 'Unnamed: 61'], axis=1)
data.dtypes
```

```python
target = all_games_df[['Points', 'Opponent Points']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 13)
X_train
```

```python
kn_model = KNeighborsRegressor()
# fit model
kn_model.fit(X_train, y_train)
# make a prediction
# yhat = model.predict

kn_model.score(X_train, y_train)

kn_model.score(X_test, y_test)

kn_model.predict(X_test)
```

```python
kn_model_points = KNeighborsRegressor()

#model_points = KNeighborsRegressor()

kn_model_points.fit(X_train_points, y_train_points)
kn_model_points.score(X_train_points, y_train_points)

kn_predict_points = kn_model_points.predict(games_today_clean)
kn_predict_points
```

```python
kn2_model_points = KNeighborsRegressor(n_neighbors=4, weights='distance', leaf_size=25)

#model_points = KNeighborsRegressor()

kn2_model_points.fit(X_train_points, y_train_points)
kn2_model_points.score(X_train_points, y_train_points)

kn2_predict_points = kn2_model_points.predict(games_today_clean)
kn2_predict_points
```

```python
kn3_model_points = KNeighborsRegressor(n_neighbors=5, weights='distance', leaf_size=25)
kn3_model_points.fit(X_train_points, y_train_points)
kn3_model_points.score(X_train_points, y_train_points)

kn3_predict_points = kn3_model_points.predict(games_today_clean)
kn3_predict_points
```
